```python
#Importando o datetime
#Heitor de Araujo Aragao
import datetime
#Menu
while True:
  try:
    opcao = int(input("---------------------------------------------------\n1- Digite no diario\n2- Leia o diario\n3- Feche o diario\n4- Apagar o conteudo do diario\n---------------------------------------------------\nEscolha uma opcao: "))
    try:
      if opcao == 1:
        print("---------------------------------------------------")
        try:
          with open("diario.txt", "a") as diario:
            x = datetime.datetime.now()
            hora = x.strftime("%d/%m/%Y %H:%M:%S")
            escrita = input("Sobre o que voce quer falar agora?: ")
            diario.write(f"[{hora}] {escrita}\n")
        except:
          print("O arquivo nao existe")
      elif opcao == 2:
        print("---------------------------------------------------")
        try:
          with open ("diario.txt", "r") as leitura:
            conteudo = leitura.read()
            if conteudo == (""):
              print("Voce ainda nao escreveu no diario")
            else:
              print(conteudo)
        except:
          print("O arquivo nao existe")

      elif opcao == 3:
        print("---------------------------------------------------")
        print("TERMINANDO O PROCESSO")
        break
#criei uma opcao para apagar o conteudo do diario
      elif opcao == 4:
        print("---------------------------------------------------")
        try:
          with open("diario.txt", "w") as apagar:
            apagar.write("")
            print("DELETANDO")
        except:
          print("O arquivo nao existe")
      else:
        print("---------------------------------------------------")
        print("Opcao invalida, escolha um numero de 1 a 4")
    except:
      print("---------------------------------------------------")
      print("ERRO")
  except:
    print("---------------------------------------------------")
    print("Escolha um numero como opcao")
```
